```
data=[[None,"Micheal"],[30,"Andy"],[19,"Justin"]]
```

```
from pyspark.sql.types import *
```

```
col=StructType([StructField("Age",IntegerType(),True),StructField("Name",StringType(),True)])
```

```
df2=spark.createDataFrame(data,col)
```

```
df2.show()
```

```
df2.printSchema()
```

```
df2.describe().show()
```

```
type(df2["Age"])
```
