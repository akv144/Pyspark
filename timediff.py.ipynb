```
dates = ['2019-07-01 12:01:19.111',
    '2019-06-24 12:01:19.222',
    '2019-11-16 16:44:55.406',
    '2019-11-16 16:50:59.406']
```

```
dates
```

```
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
dates=list(zip(dates))
```

```
df=spark.createDataFrame(dates,["input_timestamp"])
```

```
display(df)
```

```
df=df.withColumn("curr_ts",current_timestamp())
```

```
display(df)
```

```
df.printSchema()
```

```
df=df.withColumn("input_timestamp",to_timestamp(df.input_timestamp))
```

```
df.printSchema()
```

```
df=df.withColumn("diff(sec)",col("curr_ts").cast("int")-col("input_timestamp").cast("int"))
```

```
df.show()
```

```
df=df.withColumn("diff(Min)",col("diff(sec)")/60)
```

```
df.show()
```

```
df=df.withColumn("diff(hrs)",col("diff(sec)")/(60*60))
```

```
df.show()
```

```
df=df.withColumn("diff(days)",col("diff(sec)")/(60*60*24))
```

```
display(df)
```
