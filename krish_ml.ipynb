```
df=spark.read.format("csv").option("header",True).option("delmiter",",").option("inferSchema",True).load("/FileStore/tables/tips.csv")
```

```
df.show()
```

```
df.printSchema()
```

```
df.columns
```

```
#handling categorial features

from pyspark.ml.feature import StringIndexer
```

```
indexer=StringIndexer(inputCol="sex",outputCol="sex1")
```

```
df_r=indexer.fit(df).transform(df)
```

```
df_r.show()
```

```
indexer=StringIndexer(inputCols=['smoker', 'day', 'time'],outputCols=['smoker1', 'day1', 'time1'])
```

```
df_r=indexer.fit(df_r).transform(df_r)
```

```
df_r.show()
```

```
from pyspark.ml.feature import VectorAssembler
```

```
df_r.columns
```

```
f=VectorAssembler(inputCols=["tip",'size','sex1','smoker1','day1','time1'],outputCol="Ind")
```

```
out=f.transform(df_r)
```

```
finaldf=out.select("Ind",'total_bill')
```

```
finaldf.count()
```

```
from pyspark.ml.regression import LinearRegression
```

```
traindf,testdf=finaldf.randomSplit([0.75,0.25])
```

```
traindf.count()
```

```
testdf.count()
```

```
model=LinearRegression(featuresCol="Ind",labelCol='total_bill')
```

```
regressor=model.fit(traindf)
```

```
regressor.coefficients
```

```
regressor.intercept
```

```
#prediction
pred=regressor.evaluate(testdf)
```

```
pred.predictions.show()
```

```
pred.r2
```

```
pred.meanSquaredError
```

```
pred.meanAbsoluteError
```
